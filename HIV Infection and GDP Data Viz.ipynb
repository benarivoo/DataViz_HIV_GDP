<h2> 2.b. Gross Domestic Production Data Data Cleaning

```python
column_names = list(df_gdp.columns.values)
```

```python
print(column_names)
```

Let's drop the unnecessary columns

```python
df_gdp
```

```python
df_gdp = df_gdp.drop(['Series Name', 'Series Code'], axis = 1)
```

Now, I will drop the unneccary rows, which are row 217 until 221

```python
df_gdp = df_gdp.drop(df_gdp.index[217:])
```

I want to make the structure of the GDP data to be similar with the HIV infection data, in which I will have a column that has the Year data, instead having the years as columns.

```python
df_gdp = df_gdp.set_index(['Country Name','Country Code'])
```

```python
df_gdp
```

```python
df1 = df_gdp.stack()
```

```python
df1 = df1.to_frame() #Because .stack() output is a Series
```

```python
df1
```

Because the missing GDP data are filled with "..", I will change it to 0 instead of NULL because we do not want to have NULL for our 

```python
df1[df1[0] == ".."] = None
```

```python
df1 = df1.fillna(0)
```

```python
df1
```

```python
df1 = df1.astype(float)
```

```python
df1 
```

```python
df1 = df1.reset_index()
```

```python
df1
```

```python
df1.rename(columns={"level_2": "Year"}, inplace=True)
```

```python
df1.rename(columns={0: "GDP in bill USD"}, inplace=True)
```

```python
df1["GDP in bill USD"] = df1["GDP in bill USD"]/1000000000
```

```python
df1
```

Now, I want to change the data in the "Year" column into integer

```python
df1["Year"] = df1["Year"].str[:4]
```

```python
df1["Year"] = df1["Year"].astype(int)
```
